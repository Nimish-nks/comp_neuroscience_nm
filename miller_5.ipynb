```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
n_cells= 60
grp=np.random.randint(1,5,size=60,dtype=int)
```

Connectivity matrix

```python
C=np.zeros((n_cells,n_cells),dtype=int)
```

```python
for i in range(n_cells):
    for j in range(n_cells):
        if i!=j: # avoid self connnections
            if grp[i] == grp[j]: # in same grp
                C[i,j]=np.random.rand()<0.5
            else: # different group
                C[i,j]=np.random.rand()<0.1
print(f'connectivity matrix {C}') 
        
```
