# Linear Algebra

vector dimension: Number of elements in ordered array  
vector length  
Unit vector  


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
v=np.array([4,1])
```

```python
# length of the vector
length = np.linalg.norm(v)
```

```python
# unit vector
unit_v = v/length
```

```python
print(f"Vector: {v}")
print(f"Length of vector: {length}")
print(f"Unit vector: {unit_v}")
```
