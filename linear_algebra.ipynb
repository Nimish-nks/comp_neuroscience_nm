```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
v=np.array([4,1])
```

```python
# length of the vector
length = np.linalg.norm(v)
```

```python
# unit vector
unit_v = v/length
```

```python
print(f"Vector: {v}")
print(f"Length of vector: {length}")
print(f"Unit vector: {unit_v}")
```

```python
# plot vector function
def plot_vector(x,col):
	origin=np.array([0,0])
	plt.quiver(*origin, *x, angles='xy', scale_units='xy', scale=1,color=col)
	plt.xlim(-5,4)
	plt.ylim(-5,4)
	plt.grid()
	
```

```python
x= np.array([3,1])
y = np.array([-1,2])

plot_vector(x,'r')
plot_vector(y,'g')

a=-1/4

b=-1/2
z=a*x+b*y
plot_vector(z,'b')
```

Span of set of vectors: It is the set of all possible linear combinations of those vectors  
A set of vector is linearly dependent if one can be writtena s linear combination of others

```python

from mpl_toolkits.mplot3d import Axes3D

def plot_3d_vector(x, col):
    origin = np.array([0, 0, 0])
    
    # Create a 3D subplot using plt
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    
    # Plot the vector as an arrow
    ax.quiver(origin[0], origin[1], origin[2],
              x[0], x[1], x[2],
              color=col)
    
    # Set axis limits
    ax.set_xlim([-5, 5])
    ax.set_ylim([-5, 5])
    ax.set_zlim([-5, 5])
    
    # Labels
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')




```

```python
a=np.array([4,2,0])
b=np.array([1,0,3])
c=np.array([5,2,3])
d=np.array([3,3,2])

plot_3d_vector(a,'c')
plot_3d_vector(b,'r')
plot_3d_vector(c,'b')
plot_3d_vector(d,'k')
```

Span of a,b,c,d is all 3D space  
Span of a,b,c is 2d plane through 3d space  
Span of a is 1d line through 3D space  
Span of a and b is a 2d plane through 3D space
